Next, lets create a foor loop that checks if the sum of the columns separated by gender is equal to the "Total" columns.


```python
#function for checking if the sum of two columns divided by gender is equal to the "Total column"
def checkTotal(*args):
#the first argumento must be the name of the dataframe
    datFrame = args[0]
    columns = args[1:]
    columnIndex = 0
    columnSum = 0
    columnTot = 0
    colName = ""
    counter = 0
    for i in columns:
        columnTot = datFrame[columns[counter]].sum()
        columnIndex = datFrame.columns.get_loc(i)
        columnSum = datFrame.iloc[:, columnIndex + 1].sum() + datFrame.iloc[:, columnIndex + 2].sum()
        check = print( columns[counter], columnTot, ", ",datFrame.columns[columnIndex + 1], " + ",datFrame.columns[columnIndex + 2], " = ", columnSum, " --> ", columnTot == columnSum )
        counter = counter + 1
    return None 
```
